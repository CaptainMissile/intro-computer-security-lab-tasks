````python
def juxtapose_letter_freq(cipher_text):
    corpus = cipher_text.translate(str.maketrans('', '', ".–-,?'! \n"))
    ln = len(corpus)
    
    cipher_letter_freq = dict()

    for letter in corpus:
        cipher_letter_freq[letter] = cipher_letter_freq.get(letter, 0) + 1

    for key in cipher_letter_freq:
        cipher_letter_freq[key] = (cipher_letter_freq[key] / ln) * 100

    sorted_eng_lang_letter_freq = sorted(eng_lang_letter_freq.items() , key = lambda x:x[1], reverse=True)
    sorted_cipher_letter_freq = sorted(cipher_letter_freq.items() , key = lambda x:x[1], reverse=True)

    print('IN CIPHER         ENG LANG')
    print('---------         ---------')
    for first,second in zip(sorted_cipher_letter_freq, sorted_eng_lang_letter_freq):
        print(f'{first[0]}-{first[1]:.3}          {second[0]}-{second[1]}')
        
        
        
def replace_cipher_letters(cipher_text, replacing_dict):
    plain_text = ""
    for letter in cipher_text :
        plain_text += replacing_dict.get(letter, letter)

    print('\nPLAIN TEXT:\n\t',plain_text)
````
